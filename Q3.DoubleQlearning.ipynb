```python
import gym
import numpy as np
import random
```

```python
env = gym.make('FrozenLake-v1')
```

```python
Q1 = {}
Q2 = {}
Q  = {}
for s in range(env.observation_space.n):
    for a in range(env.action_space.n):
        Q1[(s,a)] = 0.0
        Q2[(s,a)] = 0.0
        Q[(s,a)]  = 0.0
```

```python
def epsilon_greedy(state, epsilon):
    if random.uniform(0,1) < epsilon:
        return env.action_space.sample()
    else:
        return max(list(range(env.action_space.n)), key = lambda x:
            Q1[(state,x)]+Q2[(state,x)])

```

```python
alpha = 0.85
gamma = 0.90
epsilon = 0.8
num_episodes = 100000
num_timesteps = 5000

for i in range(num_episodes):
    s, info = env.reset()
    for t in range(num_timesteps):
        a = epsilon_greedy(s,epsilon)
        s_, r, done, _, _ = env.step(a)
        a_1 = np.argmax([Q1[s_, a] for a in range(env.action_space.n)])
        a_2 = np.argmax([Q2[s_, a] for a in range(env.action_space.n)])
        if random.random() < .5:
            Q1[(s,a)] += alpha * (r + gamma * Q2[(s_,a_1)]-Q1[(s,a)])
        else:
            Q2[(s,a)] += alpha * (r + gamma * Q1[(s_,a_2)]-Q2[(s,a)])
        s = s_
        if done:
            break
for key, value in Q.items():
    Q[key] = (Q1[key]+Q2[key])/2
    print(key)
    print(Q[key])
```

```python
optimal_policy = np.zeros(16)
for i in range(16):
    best_value_i = 0
    for j in range(4):
        if best_value_i < Q[(i, j)]:
            best_value_i = Q[(i, j)]
            optimal_policy[i] = j
print(optimal_policy)

optimal_policy = np.zeros(16)
for i in range(16):
    best_value_i = 0
    for j in range(4):
        if best_value_i < Q1[(i, j)]:
            best_value_i = Q1[(i, j)]
            optimal_policy[i] = j
print(optimal_policy)

optimal_policy = np.zeros(16)
for i in range(16):
    best_value_i = 0
    for j in range(4):
        if best_value_i < Q2[(i, j)]:
            best_value_i = Q2[(i, j)]
            optimal_policy[i] = j
print(optimal_policy)
```
